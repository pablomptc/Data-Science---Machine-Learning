# DESCRIPTIVE ANALYSIS: DISPERSION AND CORRELATION MATRIX

**Dispersion matrix (seaborn)**

```python
cols = ["CRIMINALITY", "ZN", "INDUSTRIAL", "CHARLES_RIVER",
              "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO",
              "B", "LSTAT", "MEDV"]

sns.pairplot (df[cols], height = 2.5)
plt.show ()
```

# PREDICTIVE ANALYSIS

### MULTIPLE LINEAR REGRESSION MODEL

```python
X = df.iloc [:, :-1].values # Select all the dimmensions except MEDV
y = df ["MEDV"].values # Both X and y needs to be converted in array through .values


X_train, X_test, y_train, y_test = train_test_split (X, y, test_size = 0.3, random_state = 0) 
# test size indicates that 30% of the data will be coverted to test. 

slr = LinearRegression()
slr.fit (X_train, y_train)
y_train_pred = slr.predict (X_train)
y_test_pred = slr.predict (X_test)
```

**Train and test errors (MSE)**

```python
print ("MSE train: %.3f, test: %.3f" % (mean_squared_error (y_train, y_train_pred),
                                       mean_squared_error (y_test, y_test_pred)))
```

```python
# There is something that we have to take into acocunt: if we vary the test_size, the errors change completely.
# In some cases the difference between the two errors is so big that it is an indicator of overfit, but changing
# the test_size sometimes the difference is not big so it is not overfit. 
```

**Coefficient of determination**

```python
# The coefficient of determination is an standardized version of MSE. It's another way to measure the performance.
```

```python
print ("R^2 train: %.3f, test: %.3f" % (r2_score (y_train, y_train_pred),
                                       r2_score (y_test, y_test_pred)))
```
